```python
vec_a = np.array([1,2,3,4])
vec_b = np.array([6,7,8,9])
print(f'''Vector a: {vec_a}
Vector b: {vec_b}
Matriz: 
{vec_a.reshape(2,2)}''')
```

```python
# Operacin element-wise
print(f'''Suma
Operacin element-wise: {vec_a} + {vec_a} = {vec_a + vec_b}''')
```

```python
# Operacin tensor-wise
print(f'''Producto punto
Operacin tensor-wise: {vec_a} * {vec_b} = {np.dot(vec_a, vec_b)}''')
```

```python
from numpy import tensordot
# Operacin tensor-wise
print('Producto matricial')
tensorwise = np.tensordot(vec_b, vec_a, axes = 0)
print('Operacin tensor-wise con la matriz:\vec_b (x) vec_a = \n{0}'.format(tensorwise))
```
