#### AdaBoost: Boosting Adaptativo
Por motivos pedag√≥gicos, la exposici√≥n del algoritmo se concentra en el caso de
clasificaci√≥n. Para aquellos casos donde se entrenan regresores d√©biles, la intuici√≥n es
similar.

El algoritmo m√°s conocido de Boosting es AdaBoostM.1 de Freund y Schapire (1997).
Consideremos el siguiente problema de clasificaci√≥n binario, donde nuestro vector objetivo
se expresa como ùë¶ œµ ùëå = {‚àí 1, 1}.

Dado un conjunto de atributos ùëã un clasificador espec√≠fico h(X) produce una predicci√≥n en
una de las dos clases. Podemos obtener la tasa de error a lo largo de todos los
clasificadores mediante:

$$ \bar \varepsilon_i = \dfrac{\sum^{N}_{i=1 \hspace{2mm}I(y_i \neq h(x_i))}}{N} $$ 

Y estimar la esperanza del error en el conjunto de clasificadores como:

$$ E_{XY} I(y \neq h(X)) $$

Los algoritmos de Boosting hacen uso de clasificadores d√©biles. Un clasificador d√©bil es
aqu√©l cuyas prediciones son marginalmente mejores que la de un clasificador aleatorio.
Para efectos pr√°cticos, Freund y Schapire (1997) consideran que un √°rbol de decisi√≥n con un
solo nodo de partici√≥n es el punto de partida m√°s aceptable para generar un clasificador
d√©bil.

El objetivo de Boosting es aplicar de forma secuencia un algoritmo de clasificaci√≥n d√©bil de
manera iterativa. As√≠ generamos una serie de clasificadores d√©biles de manera:

$$ H_m(X) = 1,2, ..., M $$

La predicci√≥n de clases para nuevas observaciones se realiza mediante mayor√≠a simple de
votos ponderados de la siguiente forma:

$$ H_m  = sign (\sum^{M}_{i=1} a_m h_m(x))$$

Los votos de cada clasificador $h_m$ se pondera mediante $a_m$ que permiten ajustar la 
contribuci√≥n de cada clasificador en el ensamble.

!['ciclo adaboost'](fotos/adaboost_ciclo.jpg)

El primer paso es la inicializaci√≥n de los ponderadores observacionales. Cada observaci√≥n
ùë•ùëñ ‚àà ùëãtendr√° un ponderador œâùëñ.

En la primera iteraci√≥n, el ponderador asignar√° un peso equiprobable de forma ùë§ùëñ = 1/ùëÅ
donde N es la cantidad de observaciones.

Para cada clasificador d√©bil en nuestro ensamble ùëö ‚àà ùëÄ:

$$ \bar \varepsilon_i = \dfrac{\sum^{N}_{i=1 \hspace{2mm}\omega I(y_i \neq h(x_i))}}{\sum^{N}_{i=1}\omega_i} $$ 

Donde sumamos y ponderamos aquellas observaciones predichas incorrectamente:

$$ (I(y_i \neq h_m(x_i))) $$

Con la tasa de error espec√≠fica del clasificador d√©bil, estimamos la chance del error con:

$$ a_m = log ((1- \varepsilon_m)/ \bar\varepsilon_m) $$

√âsta chance permitir√° ponderar la contribuci√≥n de cada clasificador. El objetivo de esta fase
es asignar un mayor peso a aquellas observaciones incorrectas, forzando la identificaci√≥n
de clasificadores correctos para √©stas.

- Antes de proseguir con el siguiente clasificador d√©bil, se actualiza el peso
  observacional ùë§ùëñ‚Üê ùë§ùëñ ¬∑ ùëíùë•ùëù [Œ±ùëö ¬∑ùêº(ùë¶ùëñ ‚â† ‚Ñé . ùëö (ùë•ùëñ)] ‚àÄùëñ ‚àà N.

- Resumir el proceso de cada clasificador d√©bil mediante $ùêª(ùë•) = ùë†ùëñùëîùëõ(\sum^{M}_{i=1} a_m h_m (x))$

Las modificaciones en cada paso de boosting consisten en aplicar ponderadores
ùë§1, ùë§2, ¬∑¬∑¬∑, ùë§ùëõ a cada observaci√≥n en el conjunto de entrenamiento. En la primera iteraci√≥n
todos los ponderadores tienen un peso equiprobable de ùë§ùëñ = 1/ùëÅ de tal manera que en la
primera iteraci√≥n el modelo entrenado no dista de Bagging o Random Forest.

Para las iteraciones posteriores ùëö ‚àà ùëÄ los ponderadores de observaciones se modifican y el
algoritmo de clasificaci√≥n se vuelve a aplicar a las observaciones ponderadas. En cada paso
m, las observaciones clasificadas incorrectamente por el clasificador anterior ‚Ñéùëö ‚àí 1(ùë•)
aumentan sus ponderadores, mientras que las observaciones clasificadas correctamente
disminuyen su ponderador.

En la medida que las iteraciones aumentan, las observaciones que son dif√≠ciles de clasificar
reciben una mayor influencia. En cada sucesi√≥n, los clasificadores se ven forzados a
concentrarse en las clasificaciones incorrectas.

#### Entrenando un clasificador Adaptativo con sklearn.ensemble.AdaBoostClassifier.
La implementaci√≥n del algoritmo se encuentra en
sklearn.ensemble.AdaBoostClassifier. De manera adicional, importamos las
funciones cl√°sicas para generar muestras de validaci√≥n y entrenamiento, as√≠ como un
reporte de m√©tricas de clasificaci√≥n. Los pasos para implementarlo ser√°n los mismos pasos
de cualquier algoritmo sklearn.

- from sklearn.model_selection import train_test_split
- from sklearn.metrics import classification_report, accuracy_score
- from sklearn.tree import DecisionTreeClassifier
- from sklearn.ensemble import AdaBoostClassifier

```python
X_train, X_test, y_train, y_test = train_test_split(df.loc[:, 'word_freq_make':'capital_run_length_total'], 
                                                    df['spam'], test_size=.33, random_state = 15820)
```

Adaptative Boosting aprovecha m√∫ltiples clasificadores d√©biles, donde su variante m√°s
com√∫n es un √°rbol de decisi√≥n con una profundidad m√≠nima. En la literatura se conoce como
una "decision stump". Si bien este es el modelo que se implementa por defecto en la clase
AdaBoostClassifier, generamos un objeto con estas caracter√≠sticas por motivos
expositivos.

En esta primera iteraci√≥n, mantendremos la cantidad de hiper par√°metros de
AdaBoostClassifier al m√≠nimo. Esto implica definir una semilla pseudoaleatoria para
facilitar la replicaci√≥n de resultados, as√≠ como un entrenamiento con decision stump con
fines comparativos.

```python
# Entrenamos el decision stump
decision_stump = DecisionTreeClassifier(max_depth=1, random_state = 15820).fit(X_train, y_train)
# Entrenamo el clasificador con adaptive boosting
adaboost_classifier = AdaBoostClassifier(base_estimator = decision_stump, random_state = 15820).fit(X_train, y_train)
```

Para efectos pr√°cticos, guardaremos las clases predichas y la probabilidad de clase
predicha para el modelo AdaBoost.

```python
# guardamos las predicciones de ambos modelos
tree_y_hat = decision_stump.fit(X_train, y_train).predict(X_test)
adaboost_y_hat = adaboost_classifier.fit(X_train, y_train).predict(X_test)
adaboost_pr = adaboost_classifier.fit(X_train, y_train).predict_proba(X_test)
```

```python
print(f'''Decision Stump:
{classification_report(y_test, tree_y_hat, digits = 3)}
Adaptative Boosting Classifier:
Cantidad de Estimadores: {adaboost_classifier.n_estimators}
{classification_report(y_test, adaboost_y_hat, digits = 3)}
''')
```

Inspeccionado los resultados del decision stump, observamos que el modelo tiene un
desempe√±o 29% superior que un clasificador d√©bil (medido con f1-score). √âsto es bastante
bueno, dado que no estamos forzando las particiones del √°rbol en gran manera. A√∫n as√≠,
cabe destacar que el desempe√±o del modelo es mejor en la predicci√≥n de aquellos mails
que no son spam.

Por defecto, al entrenar un modelo con AdaBoost y un total de 50 √°rboles con un m√≠nimo de
profundidad, el desempe√±o es un 43% superior que un clasificador aleatorio y un 15%
superior que un decision stump (medido con f1-score). Otro aspecto a considerar es que el
desempe√±o condicional a cada clase predicha disminuy√≥ sus brecha, teniendo un
desempe√±o mucho m√°s homog√©neo a lo largo de todas las clases.

#### La ponderaci√≥n observacional en Adaptive Boosting

Al inicio de la lectura mencionamos que una de las virtudes del algoritmo AdaBoost es la
capacidad de ponderar de manera secuencial las observaciones correctas e incorrectas.
Para lograr esto, debemos extraer las predicciones en cada una de las iteraciones del
modelo. La primera parte implica evaluar de forma parcelada el comportamiento de las
clases predichas en cada estimador. Esto lo logramos con la opci√≥n
sklearn.ensemble.AdaBoostClassifier().staged_predict_proba(), donde
ingresamos la matriz de atributos de validaci√≥n.


```python
# iniciamos un array vac√≠o
store_pr = []
for i in adaboost_classifier.staged_predict_proba(X_test):
    # guardamos los resultados de cada estimador
    store_pr.append(i)
```

La lista store_pr registra las predicciones para los 1519 registros de validaci√≥n. Nuestro
experimento es evaluar el patr√≥n de predicci√≥n de las probabilidades de clase en un
conjunto aleatorio de registros.

```python
# fijamos el tama√±o del gr√°fico
plt.figure(figsize = (10, 10))
# seteamos una semilla pseudo aleatoria para replicaci√≥n
np.random.seed(11238)
# Para cada uno de los n√∫meros aleatorios a elegir
for index, value in enumerate(np.random.randint(1, 50, size = 9)):
    plt.subplot(3, 3, index + 1)
    # perfilamos el comportamiento
    mlc = afx.adaboost_classification_behavior(store_pr, adaboost_pr, value)
    plt.title(f'Observaci√≥n: {value} / Clase: {mlc}')
    plt.ylim((0, 1))
    plt.ylabel('Probabilidad de clase')
    plt.xlabel('Secuencia de Estimadores')
    plt.tight_layout()
```

Algunos aspectos a considerar en la figura creada:

1. A grandes rasgos, la principal caracter√≠stica de las probabilidades predichas en cada
   modelo es su tendencia a estabilizarse en la medida que la secuencia de
   estimadores aumenta.
2. Existen casos como 37, donde en las primeras iteraciones las probabilidades
   predichas son inestables. Algunas veces se posicionan con valores altos (situaci√≥n
   que ocurre en los primeros iteradores). Casos como este son la principal motivaci√≥n
   de AdaBoost. Se observa un aumento substancial en la ponderaci√≥n asignada,
   representada con una curva m√°s pronunciada que se suaviza de forma secuencial en
   la medida que aumentan los estimadores.
3. Otros casos, como la observaci√≥n 6 presenta un comportamiento m√°s estable, que
   significa que los pesos observaciones tuvieron un menor rol.

##### Comportamiento de AdaBoost en la exactitud general

Otra caracter√≠stica deseable de Adaptive Boosting es el hecho que presenta muy buenos
resultados sin necesidad de modificar hiperpar√°metros. Para ello graficaremos la exactitud
en las predicciones de clase a lo largo de la secuencia de estimadores. Tomemos la primera
iteraci√≥n, que tiene un valor de .78. √âste es id√©ntico al valor en la exactitud del decision
stump. En las posteriores iteraciones de la secuencia, la exactitud se dispara optimiz√°ndose
en 38.


```python
store_acc = np.empty(adaboost_classifier.n_estimators)
for index, value in enumerate(adaboost_classifier.staged_score(X_test, y_test)):
    store_acc[index] = value
# extraemos la posici√≥n √≥ptima de estimadores
signal_optim = np.where(store_acc == np.max(store_acc))[0][0]
# graficamos
plt.plot(store_acc, '.-', lw = 1, label = 'Puntaje en estimador')
plt.axvline(signal_optim, color = 'dodgerblue', lw = 1, linestyle = '--', label = f'N√∫mero eficiente de estimadores: {signal_optim}')
plt.xlabel('Secuencia de Estimadores')
plt.ylabel('Exactitud en Prueba')
plt.legend(loc = 'center left', bbox_to_anchor = (1, .5));
```

#### Tasa de Aprendizaje

La tasa de aprendizaje (o learning_rate en la funci√≥n
sklearn.ensemble.AdaBoostClassifier) permite regularizar la influencia de cada
clasificador basal en los ponderadores observacionales œâùëñ.

Los valores bajos de la tasa de aprendizaje conllevan a que la actualizaci√≥n de los
ponderadores observaciones (ya sea incremento o decrecimiento) sea m√°s lenta. En este
escenario, el tiempo de entrenamiento del modelo ser√° m√°s lento. Lo inverso sucede cuando
la tasa de aprendizaje es sustancialmente alta.

```python
afx.learning_rate_hyperparams()
```

El efecto de la tasa de aprendizaje se ejemplifica con la figura
afx.learning_rate_hyperparams. Uno de los primeros puntos a considerar es el hecho
que el comportamiento de las tasas de error condicional a la tasa de aprendizaje es
relativamente similar en el rango de evaluaci√≥n. De hecho, cuando evaluamos el modelo con
una tasa superior a 2.25, el comportamiento de ambas curvas se dispara en la tasa de error
y ambas son id√©nticas.

Entre el rango de 0.01 y 2.05, el comportamiento de la superficie de respuesta es
substancialmente distinto. Cuando la tasa de aprendizaje es muy baja, la soluci√≥n que
entrega es insuficiente en el conjunto de datos. Lo mismo ocurre cuando la tasa es muy alta.
Resulta que el learning rate que presente un menor error se va a asemejar a la superficie de
respuesta similar a la desarrollada con un adaboost.

```python

# imitamos el comportamiento de √°rboles d√©biles con un decision stump
tree_decision_stump = DecisionTreeClassifier(max_depth = 1).fit(X_train, y_train)
tree_model = DecisionTreeClassifier().fit(X_train, y_train)
rf_model = RandomForestClassifier(max_depth = 1).fit(X_train, y_train)
adaboost_model = AdaBoostClassifier(base_estimator = tree_decision_stump, random_state = 15820).fit(X_train, y_train)
afx.plot_roc(model = tree_decision_stump, y_true = y_test, X_test = X_test, model_label = "Decision Stump")
afx.plot_roc(model = tree_model, y_true = y_test, X_test = X_test, model_label = "Tree")
afx.plot_roc(model = adaboost_model, y_true= y_test, X_test = X_test, model_label = "AdaBoost")
afx.plot_roc(model = rf_model, y_true = y_test, X_test = X_test, model_label = "Random Forest")
plt.plot([0, 1], ls = "--")
plt.legend(loc = 'center left', bbox_to_anchor = (1, .5))
```
